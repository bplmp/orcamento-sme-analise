```python
# merge
delta = list(contratos_df.columns.values[
    np.invert(
        np.in1d(contratos_df.columns.values, empenhos_contratos_df.columns.values)
        )
    ]
)
delta.append('codContrato')
contratos_df_delta = contratos_df[delta]
temp = pd.merge(empenhos_contratos_df[['codEmpenho', 'codContrato']], contratos_df_delta, how='inner', 
         on='codContrato')
empenhos_df = pd.merge(empenhos_df, temp, how='left', 
         on='codEmpenho')
```

```python
def groupSum(df):
    terms = [
        'txtCategoriaEconomica',
        'txtDescricaoElemento',
        'txtDescricaoFonteRecurso',
        # 'txtDescricaoFuncao',
        'txtDescricaoItemDespesa',
        # 'txtDescricaoOrgao',
        'txtDescricaoPrograma',
        'txtDescricaoProjetoAtividade',
        'txtDescricaoSubElemento',
        'txtDescricaoSubFuncao',
        'txtDescricaoUnidade',
        'txtGrupoDespesa',
        'txtModalidadeAplicacao',
        'txtRazaoSocial'
    ]
    for el in terms:
        result = (df
         .groupby(el)['valTotalEmpenhado', 'valLiquidado']
         .sum().apply(lambda x: x / 1000000).sort_values('valLiquidado', ascending=False)
        )
        display(result)
    display(
    # HTML(
    df[['txtObjetoContrato']].drop_duplicates(subset='txtObjetoContrato')
        # .to_html())
    )
```

```python
def groupSumContratos(df):
    terms = ['anoExercicio',
        'txtDescricaoModalidade',
        'txtDescricaoOrgao',
        'txtObjetoContrato',
        'txtRazaoSocial',
        'txtTipoContratacao',
    ]
    for el in terms:
        result = (df
         .groupby(el)['valTotalEmpenhado', 'valLiquidado']
         .sum().apply(lambda x: x / 1000000).sort_values('valLiquidado', ascending=False)
        )
        display(result)
```

```python
# fuzz is used to compare TWO strings
from fuzzywuzzy import fuzz

# process is used to compare a string to MULTIPLE other strings
from fuzzywuzzy import process
```

```python
choices = ['fuzzy fuzzy was a bear', 'is this a test', 'THIS IS A TEST!!']
process.extract("this is a test", choices, scorer=fuzz.ratio)
```

```python
choices = unidades['nome']
match = contratos_df[contratos_df['codContrato'] == 24149]
name = match['txtObjetoContrato'].iloc[0]
print(name)
process.extract(name, choices, scorer=fuzz.ratio)
```

```python
import re
contratos_with_unidades_filter = (
    (contratos_df['txtObjetoContrato'].str.contains('\\b(cei|emei|ceu|emef|cieja)', flags=re.IGNORECASE, regex=True))
)
contratos_to_reg = contratos_df[contratos_with_unidades_filter]
unidades_nomes = unidades['nome']
for index, row in contratos_to_reg[1:100].iterrows():
    text = row['txtObjetoContrato']
    matches = process.extract(text, unidades_nomes, scorer=fuzz.partial_ratio, limit=4)
    print(text, '--------------->', matches[0][0], matches[0][1])
#     contratos_to_reg.iloc[index]['matches'] = matches
#     contratos_to_reg.iloc[index]['nearest_match'] = matches[0][0]
```
